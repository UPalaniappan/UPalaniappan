```python
import pandas as pd
import missingno as msno
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/kedeisha1/Challenges/main/coaster_db.csv')
```

```python
data.head()
```

## How many columns and rows are in the dataset?

```python
#num of rows and columns
counts=data.shape
print('Rows:'+ str(counts[0]))
print('Cloumns:'+ str(counts[1]))
```

## Is there any missing data?

```python
#Using missingno plotting the missing values
msno.matrix(data, labels=True)
```

```python
#is null function
missing=data.isnull().sum()
print(f'Missing values:\n{missing}')
```

## Display the summary statistics of the numeric columns using the describe method

```python
# The describe method returns the data of the numeric columns in the dataframe
data.describe()
```

## Rename the following columns:

```python
data = data.rename({'coaster_name':'Coaster_Name', 
                    'year_introduced':'Year_Introduced',
                    'opening_date_clean':'Opening_Date',
                    'speed_mph':'Speed_mph', 
                    'height_ft':'Height_ft', 
                    'Inversions_clean':'Inversions', 
                    'Gforce_clean':'Gforce'}, axis='columns')
```

```python
data.head()
```

## Are there any duplicated rows?

```python
# duplicated function returns true if there are any duplicated values.
duplicate = data[data.duplicated()]

print('Duplicate values:')
if duplicate.empty == True:
    print('No duplicate rows')
else:    
    print(duplicate)
```

## What are the top 3 years with the most roller coasters introduced?

```python
#group and sort
year_total=data.groupby(data['Year_Introduced']).size()
year_total=year_total.sort_values(ascending=False).head(3)
print(year_total)
```

## What is the average speed? Also display a plot to show it’s distribution.

```python
average_speed=data['Speed_mph'].mean()
print(average_speed)
```

```python
#histogram 
plt.hist(data['Speed_mph'], bins=20, color="Green")
plt.xlabel("Speed")
plt.ylabel("Count")
plt.title("Distribution of speed")
plt.grid(color="Grey")
plt.show()
```

### Explore the feature relationships. Are there any positively or negatively correlated relationships?

```python
# using corr function to return correlation of different columns

corr=data.corr()
# matshow to display the correlation
plt.matshow(data.corr())
plt.title("Correlation")
xticklabels=corr.columns.values
yticklabels=corr.columns.values
plt.show()
```

```python
sns.heatmap(corr, annot=True)
```

### Most used type of material

```python
types=data.groupby(data['Type']).size()
types=types.sort_values(ascending=False).head(5)
print(types)
```

```python

plt.pie(types, labels=types.values)
plt.legend(types.index, bbox_to_anchor=(1.5,0.5))
plt.title("Most used type of material to make roller coasters")
plt.show()
```
